```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import folium
import numpy as np
```

```python
def get_new_york_data():
    #get data contain borough, neighborhood, latitude and longitude information
    Data = "https://cocl.us/new_york_dataset"
    re = requests.get(Data).json()
    features = re['features']
    col = ['Borough', 'Neighborhood', 'Latitude', 'Longitude']
    new_york_data = pd.DataFrame(columns=col)

    for feature in features:
        borough = feature['properties']['borough']
        neighborhood = feature['properties']['name']

        lat_lon = feature['geometry']['coordinates']
        latitude = lat_lon[1]
        longitude = lat_lon[0]

        new_york_data = new_york_data.append({'Borough': borough, 'Neighborhood': neighborhood, 'Latitude': latitude, 'Longitude': longitude}, ignore_index=True)

    return new_york_data

df1 = get_new_york_data()
df1.head(10)
```

```python
col1 = ['name', 'categories'] + [col for col in df4.columns if col.startswith('location.')] + ['id']
df5 = df4.loc[:, col1]


def get_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
df5['categories'] = df5.apply(get_type, axis=1)

# clean column names by keeping only last term
df5.columns = [column.split('.')[-1] for column in df5.columns]

df5=df5[['name','categories','lat','lng','distance']]
df5
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=10) 


# food as blue 
for lat, lng in zip(df5.lat, df5.lng):
    folium.features.CircleMarker(
        [lat, lng],
        radius=10,
        color='blue',
        fill = True,
        fill_color='blue',
        fill_opacity=0.5
    ).add_to(venues_map)

# display map
venues_map
```
